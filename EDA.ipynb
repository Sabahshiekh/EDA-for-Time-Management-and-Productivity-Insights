```python
# prompt: check outliers for each features

def check_outliers_iqr(data, column):
  Q1 = data[column].quantile(0.25)
  Q3 = data[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
  return outliers

# Check outliers for each numerical feature
numerical_features = ['Age', 'Daily Work Hours', 'Daily Leisure Hours', 'Productivity Score', 'Daily Sleep Hours', 'Daily Exercise Minutes']
for feature in numerical_features:
  outliers = check_outliers_iqr(data, feature)
  if not outliers.empty:
    print(f"Outliers found for {feature}:\n{outliers}")
  else:
    print(f"No outliers found for {feature}")
```

```python
# prompt: check outliers for each column

def check_outliers_iqr(data, column):
  Q1 = data[column].quantile(0.25)
  Q3 = data[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
  return outliers

# Check outliers for each numerical feature
numerical_features = ['Age', 'Daily Work Hours', 'Daily Leisure Hours', 'Productivity Score', 'Daily Sleep Hours', 'Daily Exercise Minutes']
for feature in numerical_features:
  outliers = check_outliers_iqr(data, feature)
  if not outliers.empty:
    print(f"Outliers found for {feature}:\n{outliers}")
  else:
    print(f"No outliers found for {feature}")
```

```python
# prompt: check outliers for each column show by graph or plots

import matplotlib.pyplot as plt
import seaborn as sns

# Function to create box plots for outlier detection
def plot_outlier_boxplots(data, features):
  for feature in features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=data[feature])
    plt.title(f'Box Plot of {feature} for Outlier Detection')
    plt.show()

# Call the function to create box plots for numerical features
numerical_features = ['Age', 'Daily Work Hours', 'Daily Leisure Hours', 'Productivity Score', 'Daily Sleep Hours', 'Daily Exercise Minutes']
plot_outlier_boxplots(data, numerical_features)
```

```python
# prompt: create box plots and distribution

import matplotlib.pyplot as plt
import seaborn as sns

# Function to create box plots and distribution plots for a given feature
def plot_box_and_distribution(data, feature):
  plt.figure(figsize=(12, 4))

  # Box plot
  plt.subplot(1, 2, 1)
  sns.boxplot(x=data[feature])
  plt.title(f'Box Plot of {feature}')

  # Distribution plot (histogram)
  plt.subplot(1, 2, 2)
  sns.histplot(data[feature], kde=True)  # kde=True adds a kernel density estimate curve
  plt.title(f'Distribution of {feature}')

  plt.show()

# Call the function for each numerical feature you want to analyze
numerical_features = ['Age', 'Daily Work Hours', 'Daily Leisure Hours', 'Productivity Score', 'Daily Sleep Hours', 'Daily Exercise Minutes']
for feature in numerical_features:
  plot_box_and_distribution(data, feature)
```

```python
# prompt:  plot correlation graph for all numerical features

import matplotlib.pyplot as plt
# Calculate the correlation matrix for numerical features
correlation_matrix = data.corr()

# Create a heatmap to visualize the correlation
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
# prompt: plot bivariate graph for all feature among one another

import seaborn as sns
import matplotlib.pyplot as plt

# Create a pairplot to visualize relationships between all numerical features
sns.pairplot(data)
plt.show()
```
